````python
Ans 1 
import scipy.stats as stats

def variance_ratio_test(data1, data2):
    # Calculate the sample variances
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)
    
    # Calculate the degrees of freedom
    n1 = len(data1)
    n2 = len(data2)
    df1 = n1 - 1
    df2 = n2 - 1
    
    # Calculate the F-value
    F = var1 / var2
    
    # Calculate the p-value
    p_value = stats.f.sf(F, df1, df2)
    
    return F, p_value
````

````python
Ans 3 
import numpy as np
import scipy.stats as stats

def variance_equality_test(sample1, sample2, var1, var2, significance_level):
    # Calculate the sample sizes
    n1 = len(sample1)
    n2 = len(sample2)
    
    # Calculate the degrees of freedom
    df1 = n1 - 1
    df2 = n2 - 1
    
    # Calculate the sample variances
    sample_var1 = np.var(sample1, ddof=1)
    sample_var2 = np.var(sample2, ddof=1)
    
    # Calculate the F-value
    F = sample_var1 / sample_var2
    
    # Calculate the p-value
    p_value = stats.f.sf(F, df1, df2)
    
    # Perform two-tailed test
    p_value *= 2
    
    # Determine if null hypothesis is rejected
    reject_null_hypothesis = p_value < significance_level
    
    return F, df1, df2, p_value, reject_null_hypothesis

# Generate random samples from two normal distributions
np.random.seed(0)
sample1 = np.random.normal(loc=0, scale=1, size=100)
sample2 = np.random.normal(loc=0, scale=1, size=120)

# Known variances
var1 = 1
var2 = 1

# Significance level
significance_level = 0.05

# Perform variance equality test
F_value, df1, df2, p_value, reject_null_hypothesis = variance_equality_test(sample1, sample2, var1, var2, significance_level)

# Output the results
print("F-value:", F_value)
print("Degrees of freedom (numerator):", df1)
print("Degrees of freedom (denominator):", df2)
print("p-value:", p_value)
print("Reject null hypothesis?", reject_null_hypothesis)
````
